```python
%matplotlib notebook
import pandas as pd
```

```python
data = pd.read_csv('flowdata.csv',index_col = 0,parse_dates = True)
data.head()
```

### 按时间切片取数据源

```python
data[pd.Timestamp('2012-01-01 09:00'):pd.Timestamp('2012-01-01 19:00')]
```

```python
data[('2012-01-01 09:00'):('2012-01-01 19:00')] #直接取也是可以的
```

### 取最后10个数据

```python
data.tail(10)
```

### 取2013年的数据

```python
data['2013']
```

### 按月份区间取

```python
data['2012-01':'2012-03']
```

### 用bool类型索引判断 取出所有一月份的

```python
data[data.index.month == 1]
```

### 用bool类型索引判断 取出8点到12点的

```python
data[(data.index.hour > 8) & (data.index.hour <12)]
```

```python
data.between_time('08:00','12:00') # 也可以这样取
```

### 时间序列的重采样

```python
data.head()
```

```python
# 以天为单位计算（重采样）平均值
data.resample('D').mean().head()
```

```python
# 以天为单位计算（重采样）最大值
data.resample('D').max().head()
```

```python
# 以3天为单位重采样 平均值
data.resample('3D').mean().head()
```

```python
# 以月为单位
data.resample('M').mean().head()
```

```python
# 画一个图，注意看开头的魔法指令
data.resample('M').mean().plot()
```
