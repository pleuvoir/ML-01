```python
import pandas as pd
data = pd.DataFrame({'group':['a','a','a','b','b','b','c','c','c'],
                    'data':[4,3,2,1,12,3,4,5,7]})
data
```

### 排序

```python
data.sort_values(by=['group','data'],ascending = [False,True],inplace=True) #第一个列降序，第二个列升序
data
```

### 按照key排序排除重复的值

```python
data = pd.DataFrame({'k1':['one']*3+['two']*4,
                   'k2':[3,2,1,3,3,4,4]})
data
```

```python
data.sort_values(by='k2')
```

```python
data.drop_duplicates() #去重 默认是两列都一样才去掉
```

```python
data
```

```python
data.drop_duplicates(subset='k1') #去重，k1一样就去掉，可以看到只保留了出现的第一行记录
```

### 同类项合并

```python
data = pd.DataFrame({'food':['A1','A2','B1','B2','B3','C1','C2'],'data':[1,2,3,4,5,6,7]})
data
```

```python
def food_map(series):
    if series['food'] == 'A1':
        return 'A'
    elif series['food'] == 'A2':
        return 'A'
    elif series['food'] == 'B1':
        return 'B'
    elif series['food'] == 'B2':
        return 'B'
    elif series['food'] == 'B3':
        return 'B'
    elif series['food'] == 'C1':
        return 'C'
    elif series['food'] == 'C2':
        return 'C'
data['food_map'] = data.apply(food_map,axis = 'columns')
data
```

### 映射转换

```python
# 字典形式的定义 
food2Convert = {
    'A1':'A啊',
    'A2':'A啊',
    'B1':'B啊',
    'B2':'B啊',
    'B3':'B啊',
    'C1':'C啊',
    'C2':'C啊'
}
data['converted'] = data['food'].map(food2Convert)
data
```
