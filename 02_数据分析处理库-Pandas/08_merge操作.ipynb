```python
import pandas as pd
```

```python
left = pd.DataFrame({'key':['K0','K1','K2','K3'],
                    'A':['A0','A1','A2','A3'],
                    'B':['B0','B1','B2','B3']})
right = pd.DataFrame({'key':['K0','K1','K2','K3'],
                    'C':['C0','C1','C2','C3'],
                    'D':['D0','D1','D2','D3']})
```

```python
left
```

```python
right
```

### 按照某个具体的指标进行合并

```python
pd.merge(left,right,on='key')  #按照key
```

#### 如果有多列表头相同会怎样？

```python
left = pd.DataFrame({'key1':['K0','K1','K2','K3'],
                     'key2':['K0','K1','K2','K3'],
                    'A':['A0','A1','A2','A3'],
                    'B':['B0','B1','B2','B3']})
right = pd.DataFrame({'key1':['K0','K1','K2','K3'],
                      'key2':['K0','K1','K2','K3'],
                    'C':['C0','C1','C2','C3'],
                    'D':['D0','D1','D2','D3']})
```

```python
left
```

```python
right
```

```python
pd.merge(left,right,on='key1') 
```

注意：上面的结果我们可以发现出现了`key2_x，key2_y`这是因为它在合并的过程中发现了相同的列值，但是我们指定的键却没有它，所以为了区分就加上了_x，_y

### 按照多表头进行合并

```python
left = pd.DataFrame({'key1':['K0','K1','K2','K3'],
                     'key2':['K0','K1','K2','K3'],
                    'A':['A0','A1','A2','A3'],
                    'B':['B0','B1','B2','B3']})
right = pd.DataFrame({'key1':['K0','K1','K2','K3'],
                      'key2':['K0','K1','K2','K5'], #注意这里不一样 K5
                    'C':['C0','C1','C2','C3'],
                    'D':['D0','D1','D2','D3']})
```

```python
left
```

```python
right
```

#### 全连接

```python
pd.merge(left,right,on=['key1','key2'])
```

这里我们发现这是一个全连接，少了一行记录

```python
pd.merge(left,right,on=['key1','key2'],how='outer',indicator=True)
```

#### 左连接

```python
pd.merge(left,right,on=['key1','key2'],how='left',indicator=True)
```

#### 右连接

```python
pd.merge(left,right,on=['key1','key2'],how='right',indicator=True)
```
