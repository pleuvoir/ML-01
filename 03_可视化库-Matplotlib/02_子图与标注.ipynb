```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

### 绘制多条线

```python
f_numpy = np.arange(0,10,0.5)
plt.plot(f_numpy,f_numpy,'r--')
plt.plot(f_numpy,f_numpy**2,'bs')
plt.plot(f_numpy,f_numpy**3,'go')
```

```python
plt.plot(f_numpy,f_numpy,'r--',
        f_numpy,f_numpy**2,'bs',
        f_numpy,f_numpy**3,'go')  #也可以不用调用多次plot方法
```

### 指定线条的宽度

```python
x = np.linspace(-10,10,50)  # -10到10之间生成50个数字
y = np.sin(x) #sin函数
plt.plot(x,y,linewidth = 3.0) # linewidth 就是粗细
```

### 指定更多参数

```python
# marker 代表关键点，markerfacecolor 关键点颜色 markersize 大小
plt.plot(x,y,color='b',linestyle=':',marker = 'o',markerfacecolor='r',markersize = 10)
```

### 可以先画图再设置参数

```python
line = plt.plot(x,y)
plt.setp(line,color='b',linestyle=':',marker = 'o',markerfacecolor='r',markersize = 10, alpha = 0.1) # alpha 透明度
```

### 子图（好几副图）

```python
# 211 表示一会要画的图是2行2列的(其实就是调整子图的结构) 最后一个1表示的是子图当中的第1个图
plt.subplot(221)
plt.plot(x,y,color='r')

# 212 表示一会要画的图是2行2列的(其实就是调整子图的结构) 最后一个4表示的是子图当中的第4个图
plt.subplot(224)
plt.plot(x,y,color='b')
```

### 给图加上注释

```python
plt.plot(x,y,color='b',linestyle=':',marker = 'o',markerfacecolor='r',markersize = 10)
plt.xlabel('x:---')
plt.ylabel('y:---')
plt.title('p-title')
plt.text(0,0,'p-text') #文字加在哪个坐标
plt.grid(True) #增加网格，方便看坐标
#增加注释，箭头的，这里就是制定箭头位置宽度颜色什么的，基本上这些属性就够用了
plt.annotate('p-annotate',xy=(-5,0),xytext=(-2,0.3),arrowprops = dict(facecolor='black',shrink=0.05,headlength= 20,headwidth = 20))
```
